```python
from ctgan import CTGAN
import pandas as pd
```

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
discrete_columns = ['text', 'Pattern Category']
```

```python
ctgan = CTGAN(epochs=1)
ctgan.fit(data, discrete_columns)
```

```python
synthetic_data = ctgan.sample(1000)
```

```python
synthetic_data.head(100)
```

```python
# prompt: create sythetic unique strings for text and Pattern Category in the df data using GAN

synthetic_data = ctgan.sample(1000)
synthetic_data['text'] = synthetic_data['text'].apply(lambda x: 'synthetic_' + x)
synthetic_data['Pattern Category'] = synthetic_data['Pattern Category'].apply(lambda x: 'synthetic_' + x)
```
